```python
import numpy as np
import pandas as pd
import sklearn
from sklearn. model_selection import train_test_split
from sklearn. linear_model import LogisticRegression
from sklearn. tree import DecisionTreeClassifier
import pickle





```

```python
data = pd.read_csv("kc_house_data.csv")

#Print the CSV File

data
```

```python
# This is Supervised learning

# Get the Input and Output

# The House Price Prediction In needs input are

#Bedrooms, Bathrooms,sqft_living, sqft_lot ,floors, waterfront, sqft_above,sqft_basement,r_built

# The output need only price

#Price

dipt = data[["bedrooms","bathrooms","sqft_living","sqft_lot","floors","waterfront","sqft_above","sqft_basement","yr_built"]]

#Print the input
dipt
```

```python
dout= data[["price"]]

#print the output

dout
```

```python
x_train,x_test,y_train,y_test = train_test_split(dipt,dout,test_size=0.10,random_state=10)
```

```python
#Logistic regression algorithm used already import the header file
lr =LogisticRegression()
lr.fit(x_train,y_train)
lr.predict(x_test)
lr.score(x_test,y_test)
```

```python
ld =DecisionTreeClassifier()
ld.fit(x_train,y_train)
ld.predict(x_test)
ld.score(x_test,y_test)
```

```python
#save as module.pickle
with open ("module.pickle","wb") as f:
    pickle.dump(lr,f)
```
